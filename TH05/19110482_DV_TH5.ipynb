```python
# cac thu vien duoc su dung (neu co)

import pandas as pd
import plotly.express as px
```

```python
# load du lieu tu file csv

data = pd.read_csv('number-of-internet-users-by-country.csv')
print(data)
```

```python
# thong ke nam 2016

data_2016 = data.query('Year == 2016')
print(data_2016)
```

```python
# ve bieu do the hien plotly

fig_1 = px.choropleth(data_2016, locations = 'Code', 
                    color = 'Number of internet users (OWID based on WB & UN)', 
                    color_continuous_scale = px.colors.sequential.Plasma)

fig_1.show()
```

```python
# ve bieu do the hien plotly

fig_1.update_layout(title_text = 'Number of internet users (% population)- 2016',
                    geo = dict(projection = {'type':'equirectangular'}))

fig_1.show()
```

```python
# ve bieu do the hien choropleth

fig_2 = px.choropleth(data, locations = 'Code', color = 'Number of internet users (OWID based on WB & UN)', 
                    hover_name = 'Entity', animation_frame = 'Year',
                    color_continuous_scale = px.colors.sequential.Plasma)

fig_2.show()
```

```python
# ve bieu do the hien choropleth

fig_2.update_layout(title_text = 'Number of internet users (% population)',
                    geo = dict(projection = {'type': 'equirectangular'}))

fig_2.show()
```
