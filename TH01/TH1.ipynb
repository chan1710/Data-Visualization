```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Ta lấy data tips có sẵn của seaborn

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

### 1.Themes (set_style)

* Trong **seaborn** có sẵn 5 themes : darkgrid, whitegrid, dark, white, ticks.
* Dùng hàm set_style của thư viện để thiết lập theme

```python
sns.set_style("white");
sns.distplot(tips_df.total_bill);
```

```python
sns.set_style("whitegrid");
sns.distplot(tips_df.total_bill);
```

### 2.Despine (despine)

* Dùng hàm despine để xóa các cạnh hoặc tin chỉnh các cạnh theo ý muốn.

```python
sns.reset_orig()
sns.distplot(tips_df.total_bill);
```

```python
#Xóa spine top và spine right
sns.distplot(tips_df.total_bill);
sns.despine()
```

```python
# Tin chỉnh spine left and bottom
sns.distplot(tips_df.total_bill);
sns.despine(offset= 5, trim=True)
```

### 3.Scale (set_context)

* Thiết lập hình vẽ phù hợp với ngữ cảnh.
* Dùng hàm set_context, trong seaborn có các ngữ cảnh có sẵn: paper, notebook, talk, poster.

```python
sns.reset_orig()

#paper
sns.set_context("poster");
sns.catplot("sex", data=tips_df, kind ='count');
```

```python
#poster
sns.set_context("poster");
sns.catplot("sex", data=tips_df, kind ='count');
```

```python
#poster với font_scale = 0.5
sns.set_context("poster", font_scale = .5);
sns.catplot("sex", data=tips_df, kind ='count');
```

### 4.Color (set_palette)

```python
# RGB, SHV (SHB, SHL), yCrCb...
```

* Seaborn cung cấp một vài bảng màu, nhưng chủ yếu chia làm 3 loại chính:

* Sequentail: YlOrRd, YlOrBr
* Discrete: Spectral, RdYlGn, ...
* Diverging: Set1, Set2, Set3, ...

```python
#YlOrRd
sns.palplot(sns.color_palette("YlOrRd", 10))
```

```python
#RdYlGn
sns.palplot(sns.color_palette("RdYlGn", 10))
```

```python
#Set1
sns.palplot(sns.color_palette("Set1", 10))
```

* Dùng hàm set_palette để thiết lập màu mặc định cho các plot

```python
sns.reset_orig()
sns.barplot(x='size', y='total_bill', hue = 'time', data=tips_df);
```

```python
sns.set_palette(sns.color_palette("hls"))
sns.barplot(x='size', y='total_bill', hue = 'time', data=tips_df);
```

```python
# set palette theo ý muốn
sns.set_palette(palette={'b','g'})
sns.barplot(x='size', y='total_bill', hue = 'time', data=tips_df);
```

###5.Manage Axis plt(xlabel, ylabel, tick_param, xlim, ylim, ...)

```python
sns.reset_orig()
# set label
sns.catplot("size", data=tips_df, kind ='count');
plt.xlabel('price');
plt.ylabel('time');
plt.show()
```

```python
sns.reset_orig()
#
sns.catplot("size", data=tips_df, kind ='count');
plt.tick_params(axis='x', colors='red', direction='out')
plt.show()
```

```python
sns.reset_orig()
# Set limits
sns.catplot("size", data=tips_df, kind ='count');
plt.xlim(0,20)
plt.ylim(0,200)
plt.show()
```

### 6.Annotations plt(anotate, text, ...)

* Để annotation trên một biểu đồ hay đồ thị của seaborn thì ta sử dụng thư viện matplotlib.pyplot


```python
sns.reset_orig()
# Chú thích chữ Thu 5 ở cột Thur
sns.catplot("day", data=tips_df, kind ='count');
plt.annotate('Thu 5', xy=(2,60), xytext=(3, 70), arrowprops=dict(facecolor='red'));
```

```python
sns.reset_orig()
#Chèn công thức toán học
sns.catplot("day", data=tips_df, kind ='count');
plt.text(0.6, 40, r'$\sum_{i=0}^\infty x_i$', fontsize=20);
```

### 7.Marginal plots (joinplot)

* Marginal plot giúp chúng ta nghiên cứu mối quan hệ giữa 2 biến, Biểu đồ trung tâm biểu thị mối tương quan (correlation), các biểu đồ biên (top, right) biểu thị mật độ (distribution) của hai biến.
* seaborn có hàm jointplot.

```python
# total_bill vs tip
sns.reset_orig()
sns.jointplot(x='total_bill', y='tip',data = tips_df, kind='scatter')
```

### 8.Multi-plot grids (FacetGrid, PairGrid, subplot)

* Khi khai phá dữ liệu đa chiều không quá phức tạp, một cách tiếp cận hữu ích là vẽ nhiều đối tượng con khác nhau của tập dữ liệu cần khai phá trên cùng một hình vẽ. Kỹ thuật này đôi khi được gọi là "lattice/Trellis plotting". Nó cho phép người xem nhanh chóng trích xuất một lượng lớn thông tin từ dữ liệu phức tạp. 
* Seaborn có hàm hộ trợ FacetGrid và PairGrid, 
* Dữ liệu của bạn phải ở dạng pandas dataframe.

#### FacectGrid

```python
# Vẽ biểu đồ bar thể hiện total_bill theo từng sex theo cả từng ngày
# Khởi tạo FaccetGrid
g = sns.FacetGrid(tips_df, col="day");
```

```python
# plot
g = sns.FacetGrid(tips_df, col="day");
g.map(sns.barplot, "sex", "total_bill");

# Có thể thay plt.hist, plt.scatter, sns.regplot, sns.barplot, sns.boxplot, sns.distplot,...để test
```

```python
# Biểu đồ scatter về tip vs total_bill theo sex và smoker
g = sns.FacetGrid(tips_df, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
```

#### PairGrid

```python
# Khởi tạo và plot
g = sns.PairGrid(tips_df, hue = 'day');
g.map(sns.scatterplot);
g.add_legend()
```

```python
# Tin chỉnh
g = sns.PairGrid(tips_df, hue = 'day');
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

#### subplot

* Để vẽ nhiều biểu đồ hay đồ thị trong cùng một cửa sổ(figure) thì ta dùng subplot của matplotlib (matplotlib.pyplot)

```python
# Hai histogram (total_bill, tip) trên cùng một đồ thị
sns.distplot(tips_df.total_bill);
sns.distplot(tips_df.tip);
```

```python
# Tách thành 2 đồ thị histogram riêng biệt nhưng cùng cửa sổ
plt.subplot(1,2,1)
sns.distplot(tips_df.total_bill);
plt.subplot(1,2,2)
sns.distplot(tips_df.tip);
```
