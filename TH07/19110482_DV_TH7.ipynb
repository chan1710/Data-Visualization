```python
# cac thu vien duoc su dung

import pandas as pd
import numpy as np
import plotly.express as px
```

a) Dùng dataset co2.csv, và grapminder.csv. Thực hiện lại các bước tiền xử lý dự liệu.

```python
# load du lieu tu file co2 va grapminder

fi_co2 = pd.read_csv('co2.csv')
fi_gm = pd.read_csv('gapminder.csv')
```

```python
data_gm = fi_gm[['Country', 'region']].drop_duplicates()
data_regions = pd.merge(fi_co2, data_gm, left_on = 'country', 
                        right_on = 'Country', how = 'inner')
data_regions = data_regions.drop('Country', axis = 'columns')

new_co2 = pd.melt(data_regions, id_vars = ['country', 'region'])
columns = ['country', 'region', 'year', 'co2']
new_co2.columns = columns

data_co2 = new_co2[new_co2['year'].astype('int64') > 1963]
data_co2 = data_co2.sort_values(by = ['country', 'year'])
data_co2['year'] = data_co2['year'].astype('int64')
```

```python
data_g = fi_gm[['Country', 'Year', 'gdp', 'population', 'fertility', 'life']]
data_g.columns = ['country', 'year', 'gdp', 'population', 'fertility', 'life']
data = pd.merge(data_co2, data_g, on = ['country', 'year'], how = 'left')
data = data.dropna()
print(data)
```

b) Vẽ biểu đồ biểu thị mỗi quan hệ giữa co2 và life sử dụng scatter. Với contry ở hover tool, và slider: year. 

```python
# ve bieu do the hien quan he co2 va life

fig_co2_life = px.scatter(data, x = 'life', y = 'co2', size = 'population', 
                            color = 'region', animation_frame = 'year', 
                            animation_group = 'country', hover_name = 'country', 
                            size_max = 60)
                 
fig_co2_life.show()
```

c) Tính correlation giữa co2 vs life

```python
# tim su tuong quan giua co2 va life

np_co2 = np.array(data['co2'])
np_life = np.array(data['life'])
np.corrcoef(np_co2, np_life)
```

d) Tương tự b,c) nhưng mà cho co2 vs fertility.

```python
# ve bieu do the hien quan he giua co2 va fertility

fig_co2_fertility = px.scatter(data, x = 'fertility', y = 'co2', size = 'population', 
                                color = 'region', animation_frame = 'year', 
                                animation_group = 'country', hover_name = 'country', 
                                size_max = 60)
                 
fig_co2_fertility.show()
```

```python
# tim su tuong quan giua co2 va fertility

np_co2 = np.array(data['co2'])
np_fertility = np.array(data['fertility'])
np.corrcoef(np_co2, np_fertility)
```

e) Tạo 1 barplot visualize co2 mỗi region, mỗi year (slider : year), hover tool : country

```python
# ve bieu do the hien co2 moi region

fig_co2_region = px.bar(data, x = 'region', y = 'co2', animation_frame = 'year', 
                        animation_group = 'region', hover_name = 'country')
              
fig_co2_region.show()
```
